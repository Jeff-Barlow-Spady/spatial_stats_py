```python

def sql_type(header):
    if header in ['the_geom', 'geometry']:
        return 'GEOMETRY(MULTIPOLYGON, 4326)'
    elif header in ['longitude', 'latitude']:
        return 'DOUBLE PRECISION'
    elif header in ['centroid']:
        return 'GEOMETRY(POINT, 4326)'
    else:
        return 'DOUBLE PRECISION'

def generate_sql_create_table(headers, table_name, primary_key=None):
    columns_sql = []
    for header in headers:
        column_sql = f'"{header}" {sql_type(header)}'
        if header == primary_key:
            column_sql += ' PRIMARY KEY'
        columns_sql.append(column_sql)
    columns_sql_str = ',\n    '.join(columns_sql)
    return f'CREATE TABLE {table_name} (\n    {columns_sql_str}\n);'



```

```python
# Generating SQL for neighbourhood_characteristics table
sql_neighbourhood_indicators = generate_sql_create_table(indicator_headers, 'neighbourhood_indicators', 'neighbourhood')
sql_neighbourhood_indices = generate_sql_create_table(vi_headers, 'neighbourhood_indices', 'neighbourhood')

# Generating SQL for neighbourhood_geometry table (only spatial columns)
spatial_headers = ['neighbourhood', 'the_geom', 'centroid']
sql_neighbourhood_geometry = generate_sql_create_table(spatial_headers, 'neighbourhood_geometry', 'neighbourhood')
sql_neighbourhood_indicators_no_z = generate_sql_create_table(indicators_without_z_headers, 'non_z_indicators')
# Output the SQL statements
print("SQL for Neighbourhood Characteristics Table:\n")
print(sql_neighbourhood_indicators)
print("\nSQL for Neighbourhood Indices Table:\n")
print(sql_neighbourhood_indices)
print("\nSQL for Neighbourhood Geometry Table:\n")
print(sql_neighbourhood_geometry)
print("\nSQL for non_z")
print(sql_neighbourhood_indicators_no_z)
```
