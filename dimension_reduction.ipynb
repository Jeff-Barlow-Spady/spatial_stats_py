```python
# Assuming df is your DataFrame and columns is your list of columns

ind_columns = ['lacking_canopy_z', 'pop_density_z', 'total_pop_z',
       'homeowners_core_z', 'total_core_z', 'tennants_core_z',
       'owners_30pct_z', 'total_30pct_z', 'tennants_30pct_z',
       'tennants_subsidized_z', 'low_income_measure_z', 'low_income_cutoff_z',
       'housing_1960_z', 'housing_1961_80_z', 'new_lim_65plus_z',
       'new_lico_65plus_z', 'prop_children_z', 'gini_index_z',
       'single_parent_z', 'bottom_decile_z', 'bottom_half_z',
       'second_decile_z', 'living_alone_z', 'less_than_hs_z',
       'non_perm_resident_z', 'non_citizens_z', 'unsuitable_housing_z',
       'lico_aftertax_z', 'lim_aftertax_z', 'renters_z', 'visible_minority_z',
       'unemployed_z', 'temperature_z', 'canopy_target_z']  # replace this with your actual list of columns

def calculate_index(x):
    if x <= -2:
        return 1
    elif -2 < x <= -1:
        return 2
    elif -1 < x <= 0:
        return 3
    elif 0 < x <= 1:
        return 4
    elif 1 < x <= 2:
        return 5
    elif x > 2:
        return 6
    else:
        return 7

for column in ind_columns:
    subset_df[column + '_VI'] = subset_df[column].apply(calculate_index)
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE

# Define the target variable
target_variable = subset_df['canopy_target_z']

# Perform PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_vi)

# Perform LDA
lda = LDA(n_components=2)
df_lda = lda.fit_transform(df_vi, target_variable)

# Perform t-SNE
tsne = TSNE(n_components=2)
df_tsne = tsne.fit_transform(df_vi)

# Print the reduced dimensions
print("PCA:")
print(df_pca)
print("LDA:")
print(df_lda)
print("t-SNE:")
print(df_tsne)
```
